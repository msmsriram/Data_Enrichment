```python
from langchain.chains import create_extraction_chain
from langchain_community.document_loaders import BSHTMLLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_community import embeddings
from langchain_community.vectorstores import FAISS
```

```python

# Schema
schema = {
    "properties": {
        "university_name": {"type": "string"},
        "physical_address": {"type": "string"},
        "contact_number": {"type": "string"},
    },
    "required": ["university_name", "physical_address","contact_number"],
}

# Input
# input = """Alex is 5 feet tall. Claudia is 1 feet taller than Alex and jumps higher than him. Claudia is a brunette and Alex is blonde."""
loader = BSHTMLLoader("C:\\Users\\Barani\\Desktop\\local_ollama\\llm_testing_urls\\test_1.html",open_encoding='utf-8')
data = loader.load()

```

```python
from langchain_community.vectorstores import Chroma
```

```python
text_splitter = CharacterTextSplitter.from_tiktoken_encoder(chunk_size=7500, chunk_overlap=200)
doc_splits = text_splitter.split_documents(data)
vectorstore = Chroma.from_documents(
            documents=doc_splits,
            collection_name="rag-chroma",
            embedding=embeddings.ollama.OllamaEmbeddings(model='nomic-embed-text'),
            
        )

# retriever = vectorstore.as_retriever()
```

```python
# Run chain
llm = OllamaFunctions(model="mistral", temperature=0)
chain = create_extraction_chain(schema, llm)
chain.run(data)
```
